# Jumping Jacks

```python
%matplotlib inline

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd

from scipy.interpolate import interp1d
from scipy.signal import argrelextrema
```

```python
FILE_PATH = 'movedata/jumpingjacks/move1.csv'

df = pd.read_csv(FILE_PATH)
# df.head(2)
```

```python
# extract values
left_foot  = list(df.FootLeft_x.values)
right_foot = list(df.FootRight_x.values)
timestamps = list(df.TimeStamp.values)

# convert to floats
left_foot  = [float(l) for l in left_foot]
right_foot = [float(r) for r in right_foot]

zipped = zip(left_foot, right_foot)
# list of feet position differences
differences = [abs(z[0]-z[1]) for z in zipped]
```

# Raw Data

```python
times = [int(t) for t in timestamps[:-1]]
diffs = [float(a) for a in differences[:-1]]

plt.plot(times, diffs, '-')
plt.show()
```

## Smooting (spline interpolation of third order)

```python
f3 = interp1d(times, diffs, kind='cubic')
xnew = np.linspace(min(times), max(times), 100)

plt.plot(xnew, f3(xnew),'-')
plt.show()

# http://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.interp1d.html
```

## Peak Counting

```python
diffs = np.array(diffs)
raw_local_maxes = argrelextrema(diffs, np.greater)

print "Number of local maxima in the raw data set: "
print len(raw_local_maxes[0])
```

# Knee to Elbow

```python
FILE_PATH = 'movedata/kneetoelbow/move1.csv'

df = pd.read_csv(FILE_PATH)
# df.head(2)
```

```python
# extract values
timestamps = list(df.TimeStamp.values)

left_elbow_x = list(df.ElbowLeft_x)
left_elbow_y = list(df.ElbowLeft_y)
                    
left_knee_x = list(df.KneeLeft_x)
left_knee_y = list(df.KneeLeft_y)

right_elbow_x = list(df.ElbowRight_x)
right_elbow_y = list(df.ElbowRight_y)
  
right_knee_x = list(df.KneeRight_x)
right_knee_y = list(df.KneeRight_y)


left_knee = zip(left_knee_x, left_knee_y)
right_knee = zip(right_knee_x, right_knee_y)

left_elbow = np.array(zip(left_elbow_x, left_elbow_y))
right_elbow = np.array(zip(right_elbow_x, right_elbow_y))
```

```python
# Eucledian distances

distances_L2R = (left_knee-right_elbow)**2
distances_R2L = (right_knee-left_elbow)**2

distances_L2R = distances_L2R.sum(axis=-1)
distances_L2R = np.sqrt(distances_L2R)

distances_R2L = distances_R2L.sum(axis=-1)
distances_R2L = np.sqrt(distances_R2L)
```

```python
times = [int(t) for t in timestamps[:-1]]
L2R = [float(a) for a in distances_L2R[:-1]]
R2L = [float(a) for a in distances_R2L[:-1]]

print "Left knee - Right elbow"
plt.plot(times, L2R, '-')
plt.show()

print "Right knee - Left elbow"
plt.plot(times, R2L, '-')
plt.show()
```

## Peaks

```python
L2R = np.array(L2R)
R2L = np.array(R2L)

raw_local_mins_L2R = argrelextrema(L2R, np.less)
raw_local_mins_R2L = argrelextrema(R2L, np.less)

print "Number of local mins in the raw data set (L2R):"
print len(raw_local_mins_L2R[0])

print "Number of local mins in the raw data set (R2L):"
print len(raw_local_mins_R2L[0])
```

```python
DELTA = 0.1

peaks_L2R, bottoms_L2R = peakdet(L2R, DELTA)
peaks_R2L, bottoms_R2L = peakdet(R2L, DELTA)

print "Number of bottoms with delta applied (L2R):"
print len(bottoms_L2R)

print "Number of bottoms with delta applied (R2L):"
print len(bottoms_R2L)
```

## Smoothing

```python
f3_L2R = interp1d(times, L2R, kind='cubic')
f3_R2L = interp1d(times, R2L, kind='cubic')
xnew = np.linspace(min(times), max(times), 100)

print "After smoothing (L2R)"
plt.plot(xnew, f3_L2R(xnew),'-')
plt.show()

print "After smoothing (R2L)"
plt.plot(xnew, f3_R2L(xnew),'-')
plt.show()
```

```python
peaks_L2R, bottoms_L2R = peakdet(f3_L2R(xnew), DELTA)
peaks_R2L, bottoms_R2L = peakdet(f3_R2L(xnew), DELTA)
print "AFTER SMOOTHING"
print "Number of peaks with delta applied (L2R): "
print len(bottoms_L2R)
print "Number of peaks with delta applied (R2L): "
print len(bottoms_R2L)
```

## Rate

```python
start = times[0]
finish = times[len(times)-1]
duration = (finish-start)/1000

rate_L2R = float(len(bottoms_L2R))/duration
rate_R2L = float(len(bottoms_R2L))/duration

rate = (rate_L2R + rate_R2L)/2

print "Total time of exercise: "
print "%d seconds" % duration

print "Rate: "
print "%.2f reps/second" % rate
```

# Arm Circles

```python
FILE_PATH = 'movedata/armcircles/move1.csv'

df = pd.read_csv(FILE_PATH)
# df.head(2)
```

```python
# extract values
timestamps = list(df.TimeStamp.values)

right_wrist_x = list(df.WristRight_x)
right_wrist_z = list(df.WristRight_z)

left_wrist_x = list(df.WristLeft_x)
left_wrist_z = list(df.WristLeft_x)
```

```python
times = [int(t) for t in timestamps[:-1]]

rw_x = [float(a) for a in right_wrist_x[:-1]]
rw_z = [float(a) for a in right_wrist_z[:-1]]

lw_x = [float(a) for a in left_wrist_x[:-1]]
lw_z = [float(a) for a in left_wrist_z[:-1]]

print "Right wrist x"
plt.plot(times, rw_x, '-')
plt.show()

print "Right wrist z"
plt.plot(times, rw_z, '-')
plt.show()

print "Left wrist x"
plt.plot(times, lw_x, '-')
plt.show()

print "Left wrist z"
plt.plot(times, lw_z, '-')
plt.show()
```

# Peaks

```python
def output_result(arr):
    print "Number of local maxima in the raw data set: "
    print len(arr[0])

rw_x = np.array(rw_x)
rw_x_maxes = argrelextrema(rw_x, np.greater)
rw_z = np.array(rw_z)
rw_z_maxes = argrelextrema(rw_z, np.greater)

lw_x = np.array(lw_x)
lw_x_maxes = argrelextrema(lw_x, np.greater)
lw_z = np.array(lw_z)
lw_z_maxes = argrelextrema(lw_z, np.greater)

print "Right wrist x"
output_result(rw_x_maxes)

print "Right wrist z"
output_result(rw_z_maxes)

print "Left wrist x"
output_result(lw_x_maxes)

print "Left wrist z"
output_result(lw_z_maxes)  
```

```python
def get_peaks_delta(arr, delta=0.1):
    peaks, _ = peakdet(arr, delta)
    return peaks

def output_result_delta(arr):
    print "Number of peaks with delta applied: "
    print len(arr)
    
rw_x_peaks = get_peaks_delta(rw_x, delta=.05)
rw_z_peaks = get_peaks_delta(rw_z, delta=.05)
lw_x_peaks = get_peaks_delta(lw_x, delta=.05)
lw_z_peaks = get_peaks_delta(lw_z, delta=.05)

print "Right wrist x"
output_result_delta(rw_x_peaks)

print "Right wrist z"
output_result_delta(rw_z_peaks)

print "Left wrist x"
output_result_delta(lw_x_peaks)

print "Left wrist z"
output_result_delta(lw_z_peaks)
```

## Rate

```python
start = times[0]
finish = times[len(times)-1]
duration = (finish-start)/1000

rate_rwx = float(len(rw_x_peaks))/duration
rate_rwz = float(len(rw_z_peaks))/duration

rate_lwx = float(len(lw_x_peaks))/duration
rate_lwz = float(len(lw_z_peaks))/duration

rate_rw = (rate_rwx+rate_rwz)/2
rate_lw = (rate_lwx+rate_lwz)/2

print "Total time of exercise: "
print "%d seconds" % duration

print "Rate (right wrist): "
print "%.2f reps/second" % rate_rw

print "Rate (left wrist): "
print "%.2f reps/second" % rate_lw
```
