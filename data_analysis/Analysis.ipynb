```python
def get_df(session_id, exercise, num):
    file_path = 'data/session%d/%s%d.csv' % (session_id, exercise, num)
    return pd.read_csv(file_path)
```

```python
def show_raw_data(times, diffs):
    plt.plot(times, diffs, '-')
    plt.show()
```

```python
def get_peaks(arr, delta=0.1, show=False):
    if show:
        arr = np.array(arr)
        peaks,_ = peakdet(arr, delta)
        print "Number of peaks with delta applied: "
        print len(peaks)
        return peaks
```

```python
def get_bottoms(arr, delta=0.1, show=False):
    if show:
        arr = np.array(arr)
        _, bottoms = peakdet(arr, delta)
        print "Number of bottoms with delta applied: "
        print len(bottoms)
        return bottoms
```

```python
def get_peaks_smooth(f3, xnew, delta=0.1):
    peaks,_ = peakdet(f3(xnew), delta)
    return peaks
```

```python
def get_bottoms_smooth(f3, xnew, delta=0.1):
    _, bottoms = peakdet(f3(xnew), delta)
    return bottoms
```

```python
def smooth_the_data(times, diffs):
    f3 = interp1d(times, diffs, kind='cubic')
    xnew = np.linspace(min(times), max(times), 100)
    return f3, xnew
```

```python
def show_smooth_data(f3, xnew):
    plt.plot(xnew, f3(xnew),'-')
    plt.show()
```

```python
def get_height(df):
    
    head = np.array(df.Head_y)
    
    head = np.array(df.Head_y)
    foot_l = np.array(df.FootLeft_y)
    foot_r = np.array(df.FootRight_y)
   
    head = argrelextrema(head, np.less)[0]
    foot_l = argrelextrema(foot_l, np.less)[0]
    foot_r = argrelextrema(foot_r, np.less)[0]
    
    h = float(sum(head))/len(head)
    fl = float(sum(foot_l))/len(foot_l)
    fr = float(sum(foot_r))/len(foot_r)
    
    return h-(float(fl+fr)/2)
```

```python
def get_rate(times, peaks):
    start = times[0]
    finish = times[len(times)-1]
    duration = (finish-start)/1000
    
    rate = float(len(peaks))/duration
    
    print ""
    print "Rate: "
    print "------------------------"
    print "%.2f reps/second" % rate
    print ""
    
    return rate
```

```python
def get_average_rate(times, peaks_left, peaks_right):
    start = times[0]
    finish = times[len(times)-1]
    duration = (finish-start)/1000
    
    rate_right = float(len(peaks_left))/duration
    rate_left = float(len(peaks_right))/duration
    
    rate = (rate_right + rate_left)/2
    
    print ""
    print "Rate: "
    print "------------------------"
    print "%.2f reps/second" % rate
    print ""
    
    return rate
```

```python
def get_average_height(peaks):
    heights = []
    for ps in peaks:
        heights.append(abs(ps[1]/2))

    average_height = sum(heights)/len(heights)
    print ""
    print "Average feet height in jump:"
    print "----------------------------"
    print average_height
    print ""
    
    return average_height
    
```

```python
def get_floor(bottoms1, bottoms2):  
    floors = []
    for b in bottoms1:
        floors.append(b[1])
    for b in bottoms2:
        floors.append(b[1])
    
    floor = sum(floors)/len(floors)
    
    return floor
```

# Jumping Jacks

```python
def analyze_jumpingjacks(session_id, exercise, num, show=True):
      
    df = get_df(session_id, exercise, num)
    
    # extract values
    left_foot  = df.FootLeft_x.values
    right_foot = df.FootRight_x.values
    timestamps = df.TimeStamp.values
    
    # convert to floats
    left_foot  = [float(l) for l in left_foot]
    right_foot = [float(r) for r in right_foot]
    
    # list of feet position differences
    zipped = zip(left_foot, right_foot)
    differences = [abs(z[0]-z[1]) for z in zipped]
    
    times = [int(t) for t in timestamps[:-1]]
    diffs = [float(a) for a in differences[:-1]]
    
    f3, xnew = smooth_the_data(times, diffs)
    
    if show:
        print "Right vs Left foot spread "
        show_smooth_data(f3, xnew)
    
    p  = get_peaks(diffs, delta=0.1)
    ps = get_peaks_smooth(f3, xnew, delta=0.1)
    
    ah = get_average_height(ps)
    h = get_height(df)
    normalized = ah/h
    
    r = get_rate(times, ps)
    
    return r, normalized
```

## Squat Jumps

```python
def analyze_squatjumps(session_id, exercise, num, show=True):
        
    df = get_df(session_id, exercise, num)
    
    # extract values
    spine = df.Head_y.values
    timestamps = df.TimeStamp.values
    
    left_foot  = df.FootLeft_y.values
    right_foot = df.FootRight_y.values
    
    times = [int(t) for t in timestamps[:-1]]
    spine = [float(a) for a in spine[:-1]]
    
    leftfoot = [float(a) for a in left_foot[:-1]]
    rightfoot = [float(a) for a in right_foot[:-1]]
    
    zipped = zip(leftfoot, rightfoot)
    feet = [z[0]+z[1] for z in zipped]
    
    f3, xnew = smooth_the_data(times, feet)
    
    if show:
        print "Average feet height"
        show_smooth_data(f3, xnew)
    
    p  = get_peaks(feet, delta=0.1)
    ps = get_peaks_smooth(f3, xnew, delta=0.1)
    
    r = get_rate(times, ps)
    
    ah = get_average_height(ps)
    h = get_height(df)
    normalized = ah/h
    
    return r, normalized

```

## High Marches    

```python
def analyze_highmarches(session_id, exercise, num, show=True):
      
    df = get_df(session_id, exercise, num)
    
    # extract values
    timestamps = df.TimeStamp.values
    
    left_ankle = df.AnkleLeft_y.values
    right_ankle = df.AnkleRight_y.values
    
    times = [int(t) for t in timestamps[:-1]]

    leftankle  = [float(a) for a in left_ankle[:-1]]
    rightankle = [float(a) for a in right_ankle[:-1]]
    
    Lf3, Lxnew = smooth_the_data(times, leftankle)
    Rf3, Rxnew = smooth_the_data(times, rightankle)
    
    Lp  = get_peaks(leftankle, delta=0.1)
    Lps = get_peaks_smooth(Lf3, Lxnew, delta=0.1)
    
    Rp  = get_peaks(rightankle, delta=0.1)
    Rps = get_peaks_smooth(Rf3, Rxnew, delta=0.1)
    
    if show:
        print ""
        print "Left Ankle"
        show_smooth_data(Lf3, Lxnew)
     
        print ""
        print "Right Ankle"
        show_smooth_data(Lf3, Lxnew)
    
    Lbtms = get_bottoms_smooth(Lf3, Lxnew, delta=0.1)
    Rbtms = get_bottoms_smooth(Rf3, Rxnew, delta=0.1)
    
    floor = get_floor(Lbtms, Rbtms )
    
    lp = [abs(p[1]-floor) for p in Lps]    
    rp = [abs(p[1]-floor) for p in Rps]
    
    print ""
    print ""
    print "Average height (left ankle):"
    print "----------------------------"
    print sum(lp)/len(lp)
    print ""
    
    print "Average height (right ankle):"
    print "-----------------------------"
    print sum(rp)/len(rp)
    print ""
    
    ah = ((sum(lp)/len(lp))+(sum(rp)/len(rp)))/2
    h = get_height(df)
    normalized = ah/h
    
    ar = get_average_rate(times, Lps, Rps) 
    
    return ar, normalized
```

## Side to Side

```python
def analyze_sidetoside(session_id, exercise, num, show=True):
        
    df = get_df(session_id, exercise, num)
    
    # extract values
    left_foot  = df.FootLeft_x.values
    right_foot = df.FootRight_x.values
    timestamps = df.TimeStamp.values
    
    # convert to floats
    left_foot  = [float(l) for l in left_foot]
    right_foot = [float(r) for r in right_foot]
        
    zipped = zip(left_foot, right_foot)
    averages = [abs(z[0]+z[1]/2) for z in zipped]
    
    times = [int(t) for t in timestamps[:-1]]
    avrgs = [float(a) for a in averages[:-1]]
        
    f3, xnew = smooth_the_data(times, avrgs)
    
    if show:
        print "Average feet position"
        show_smooth_data(f3, xnew)
    
    p  = get_peaks(avrgs, delta=0.1)
    ps = get_peaks_smooth(f3, xnew, delta=0.1)
    
    r = get_rate(times, ps)
    return r
```

## Knee to Elbow

```python
def analyze_kneetoelbow(session_id, exercise, num, show=True):
    
    df = get_df(session_id, exercise, num)
    
    # extract values
    timestamps = list(df.TimeStamp.values)
    
    left_elbow_x = list(df.ElbowLeft_x)
    left_elbow_y = list(df.ElbowLeft_y)
                        
    left_knee_x = list(df.KneeLeft_x)
    left_knee_y = list(df.KneeLeft_y)
    
    right_elbow_x = list(df.ElbowRight_x)
    right_elbow_y = list(df.ElbowRight_y)
      
    right_knee_x = list(df.KneeRight_x)
    right_knee_y = list(df.KneeRight_y)
    
    left_knee = zip(left_knee_x, left_knee_y)
    right_knee = zip(right_knee_x, right_knee_y)
    
    left_elbow = np.array(zip(left_elbow_x, left_elbow_y))
    right_elbow = np.array(zip(right_elbow_x, right_elbow_y))
    
    # Eucledian distances
    distances_L2R = (left_knee-right_elbow)**2
    distances_R2L = (right_knee-left_elbow)**2
    
    distances_L2R = distances_L2R.sum(axis=-1)
    distances_L2R = np.sqrt(distances_L2R)
    
    distances_R2L = distances_R2L.sum(axis=-1)
    distances_R2L = np.sqrt(distances_R2L)
    
    times = [int(t) for t in timestamps[:-1]]
    L2R = [float(a) for a in distances_L2R[:-1]]
    R2L = [float(a) for a in distances_R2L[:-1]]
    
    L2R_f3, xnew = smooth_the_data(times, L2R)
    R2L_f3, xnew = smooth_the_data(times, R2L)
    
    if show:
        print "Left knee to Right elbow"
        show_smooth_data(L2R_f3, xnew)
        
        print "Right knee to Left elbow"
        show_smooth_data(R2L_f3, xnew)
      
    bottoms_L2R = get_bottoms_smooth(L2R_f3, xnew, delta=0.1)
    bottoms_R2L = get_bottoms_smooth(R2L_f3, xnew, delta=0.1)
    
    ar = get_average_rate(times, bottoms_L2R, bottoms_R2L)
    return ar
```

## Tuck Jumps

```python
def analyze_tuckjumps(session_id, exercise, num, show=True):
    
    df = get_df(session_id, exercise, num)
    
    # extract values
    timestamps = df.TimeStamp.values
    
    left_knee_y = df.KneeLeft_y
    right_knee_y = df.KneeRight_y
    
    zipped = zip(left_knee_y, right_knee_y)
    knees_added = [z[0]+z[1] for z in zipped]
        
    times = [int(t) for t in timestamps[:-1]]
    knees = [float(a) for a in knees_added[:-1]]
    
    f3, xnew = smooth_the_data(times, knees)
    
    if show:
        print "Average knee height"
        show_smooth_data(f3, xnew)
    
    p  = get_peaks(knees, delta=0.1)
    ps = get_peaks_smooth(f3, xnew, delta=0.1)
    
    r = get_rate(times, ps)
    return r
```

```python
def analyze_armcircles(session_id, exercise, num):
    
    df = get_df(session_id, exercise, num)
    
    # extract values
    timestamps = df.TimeStamp.values
    
    right_wrist_x = df.WristRight_x
    right_wrist_z = df.WristRight_z
    
    left_wrist_x = df.WristLeft_x
    left_wrist_z = df.WristLeft_x

    times = [int(t) for t in timestamps[:-1]]
    
    rw_x = [float(a) for a in right_wrist_x[:-1]]
    rw_z = [float(a) for a in right_wrist_z[:-1]]
    
    lw_x = [float(a) for a in left_wrist_x[:-1]]
    lw_z = [float(a) for a in left_wrist_z[:-1]]
    
    
    rwx_f3, xnew = smooth_the_data(times, rw_x)
    rwz_f3, xnew = smooth_the_data(times, rw_z)
    lwx_f3, xnew = smooth_the_data(times, lw_x)
    lwz_f3, xnew = smooth_the_data(times, lw_z)
        
    def get_peaks_delta(arr, delta=0.1):
        peaks, _ = peakdet(arr, delta)
        return peaks
    
    def output_result_delta(arr):
        print "Number of peaks with delta applied: "
        print len(arr)
        
    rw_x_peaks = get_peaks_delta(rw_x, delta=.05)
    rw_z_peaks = get_peaks_delta(rw_z, delta=.05)
    lw_x_peaks = get_peaks_delta(lw_x, delta=.05)
    lw_z_peaks = get_peaks_delta(lw_z, delta=.05)
    
    rwx_f3, xnew = smooth_the_data(times, rw_x)
    rwz_f3, xnew = smooth_the_data(times, rw_z)
    lwx_f3, xnew = smooth_the_data(times, lw_x)
    lwz_f3, xnew = smooth_the_data(times, lw_z)

    rw_x_peaks = get_peaks_smooth(rwx_f3, xnew, delta=0.05)
    rw_z_peaks = get_peaks_smooth(rwz_f3, xnew, delta=0.05)
    lw_x_peaks = get_peaks_smooth(lwx_f3, xnew, delta=0.05)
    lw_z_peaks = get_peaks_smooth(lwz_f3, xnew, delta=0.05)
    
    # Rate
    start = times[0]
    finish = times[len(times)-1]
    duration = (finish-start)/1000
    
    rate_rwx = float(len(rw_x_peaks))/duration
    rate_rwz = float(len(rw_z_peaks))/duration
    
    rate_lwx = float(len(lw_x_peaks))/duration
    rate_lwz = float(len(lw_z_peaks))/duration
    
    rate_rw = (rate_rwx+rate_rwz)/2
    rate_lw = (rate_lwx+rate_lwz)/2
    
    avrg = (rate_rw+rate_lw)/2
    
    print ""
    print "Rate (right wrist):"
    print "-----------------------"
    print "%.2f reps/second" % rate_rw
    print ""
    print "Rate (left wrist):"
    print "-----------------------"
    print "%.2f reps/second" % rate_lw
    print ""
    
    print ""
    print "Average rate:"
    print "-----------------------"
    print "%.2f reps/second" % avrg
    print ""
    
    return avrg
```
